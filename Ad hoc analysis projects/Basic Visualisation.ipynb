# Basic Visualisation 

In this analysis, we will visualise Immigration to Canada using line charts. The dataset was obtained from IBM.

```python
import numpy as np
import pandas as pd
```

```python
df_can = pd.read_excel('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.xlsx',
    sheet_name='Canada by Citizenship', skiprows=range(20), skipfooter=2)

print("Data read into a pandas dataframe!")
```

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
print('Matplotlib version: ', mpl.__version__)
```

```python
print(plt.style.available)
mpl.style.use(['ggplot'])
```

```python
haiti = df_can.loc['Haiti', years]
haiti.head()
```

```python
haiti.plot();
```

```python
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show()
```

```python
haiti.index
```

```python
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.text(2000, 6000, '2010 Earthquake')

plt.show()
```

```python
df_CI = df_can.loc[['China', 'India'], years]
df_CI.head()
```

```python
df_CI = df_CI.transpose()
df_CI.head(3)
```

```python
df_CI.plot(kind='line')

plt.title('Immigration from China & India')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show();
```
